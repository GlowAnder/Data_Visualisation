```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# question no=1
data_model=pd.read_csv('temperature_data.csv')
data_model.head(10)
```

```python
data_model['Date Time']=pd.to_datetime(data_model['Date Time'])
data_model.head(10)
```

```python
data_model['Date Time']=pd.to_datetime(data_model['Date Time'])
data_model=data_model.head(10)
x_list=list(pd.to_datetime(data_model['Date Time']))
# y_list=list(data_model['T (degC)'])
```

```python
# x=df.loc[0:10,Date Time]
x=x_list

# y=df.loc[0:10,T (degC)]
y=(data_model['T (degC)'])

fig,al=plt.subplots()

al.plot(x,y,marker='x',label='pattern')

# fig,plt.subplots()

plt.title('<<<<TEMPERATURE>>>>>')

plt.xlabel('===DATE TIME===')

plt.ylabel('====TEMPERATURE====')

plt.legend(['lines'])

plt.xticks(rotation=30)

plt.show()
```

```python
# questionno=2
df=pd.read_csv('Temp_and_rain.csv')

df=df.head(10)

x=(df['Month'])

y=(df['rain'])

colour='green','blue','black','brown','purple','pink','orange','red','grey','violet'

bars=plt.bar(x, y, color=colour)

plt.title('<<<RAINFALL MONTHLY>>>')

plt.xlabel('---MONTHS---')

plt.ylabel('---RAINFALL AMOUNT IN mm---')

plt.legend(bars,['jan','feb','march','Apr','May','June','Sept','Oct','Nov','Dec'])

plt.xticks(rotation=45)

plt.show()
```

```python
# question no 3
graded_students=pd.read_csv('Student Exam Scores.csv')

x=graded_students['MathScore']

plt.hist(x,bins=10,alpha=0.7)

graded_students=graded_students.head(40)

plt.title('<<<SCORE FREQUENCY OF STUDENTS>>>')

plt.xlabel('===SCORES===')

plt.ylabel('=== FREQUENCY===')

plt.show()
```

```python
# question number 4

df=pd.read_csv('bmi.csv')

df=df.head(35)

x=df['Height']

y=df['Weight']

plt.scatter(x,y,color='green')

plt.subplot()

plt.title('__BMI__')

plt.xlabel('__HEIGHT__')

plt.ylabel('__WEIGHT__')

plt.grid(True)

plt.show()
```
